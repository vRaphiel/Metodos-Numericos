## Analizamos los tiempos

```python
tiempos_it = []
tamanios = [20, 30, 40, 50, 100, 250, 500, 1000, 2500]

for item in range(0, 10):
    tiempos = []
    df_tiempos = pd.read_csv(F"resultados/resultado-it-{item}.csv")
    for index, row in df_tiempos.iterrows():
        path_file = row[1].split("/")
        path = F"{path_file[0]}/{path_file[1]}/edd-tiempos-{(index + 1)}"
        tiempos.append(leer_res(F"{path}{row[12]}"))
    tiempos_it.append(tiempos)
```

```python
tiempos_LU = []
tiempos_JS = []
tiempos_JM = []
tiempos_GSS = []
tiempos_GSM = []
for i in range(0, len(tiempos_it[0])):
    for j in range(0, 5):
        tiempo_minimo = tiempos_it[0][i][j]
        for k in range(0, len(tiempos_it)):
            if(tiempos_it[k][i][j] < tiempo_minimo):
                tiempo_minimo = tiempos_it[k][i][j]
                
        if j == 0: 
            tiempos_LU.append(tiempo_minimo)
        if j == 1: 
            tiempos_JS.append(tiempo_minimo)
        if j == 2: 
            tiempos_JM.append(tiempo_minimo)
        if j == 3: 
            tiempos_GSS.append(tiempo_minimo)
        if j == 4: 
            tiempos_GSM.append(tiempo_minimo)
```

```python
tiempos = [
    tiempos_LU,
    tiempos_JS,
    tiempos_JM,
    tiempos_GSS,
    tiempos_GSM
] 
```
