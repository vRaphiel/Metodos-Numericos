# Ejercicio 4a

### Matrices de múltiples tamaños

```python
df_resultados_4a = pd.read_csv("resultados/resultado.csv");
```

```python
# Obtenemos los RANDOM
df_resultados_4a_random = df_resultados_4a[df_resultados_4a["dataset"]=="RANDOM"].copy();
fig, ax = plt.subplots()
fig = sns.scatterplot(data=df_resultados_4a_random, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.xscale("log")
plt.yscale("log")
plt.savefig("graficos/4a-pivot-tridg-random.svg");
plt.show()
```

```python
# Obtenemos los RANDOM
df_resultados_4a_random = df_resultados_4a[df_resultados_4a["dataset"]=="RANDOM"].copy();
fig, ax = plt.subplots()
fig = sns.scatterplot(data=df_resultados_4a_random, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/4a-pivot-tridg-500.svg");
plt.show()
```

```python
# Obtenemos los EDD
df_resultados_4a_EDD = df_resultados_4a[df_resultados_4a["dataset"]=="EDD"].copy();
fig, ax = plt.subplots()
fig = sns.scatterplot(data=df_resultados_4a_EDD, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.xscale("log")
plt.yscale("log")
plt.xlim((10**2)*0.5, 10**4)
plt.ylim(10**-4, 10**2)
plt.savefig("graficos/4a-pivot-tridg-edd.svg");
plt.show()
```

```python
# Obtenemos los SDP
df_resultados_4a_SDP = df_resultados_4a[df_resultados_4a["dataset"]=="EDD"].copy();
fig, ax = plt.subplots()
fig = sns.scatterplot(data=df_resultados_4a_SDP, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.xscale("log")
plt.yscale("log")
plt.xlim((10**2)*0.5, 10**4)
plt.ylim(10**-4, 10**2)
plt.savefig("graficos/4a-pivot-tridg-sdp.svg");
plt.show()
```
